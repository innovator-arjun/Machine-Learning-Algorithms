```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_ads=pd.read_csv('Social_Network_Ads.csv')
```

```python
data_ads.head(20)
```

```python
X=data_ads.iloc[:,[2,3]].values
X
```

```python
y=data_ads.iloc[: , 4].values
```

```python
y
```

```python
data_ads.shape
```

```python
#100 for test set
#300 for training set
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#Feature Scalling for accurate result
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X=StandardScaler()

X_train=sc_X.fit_transform(X_train)

X_test=sc_X.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression(random_state=0)
```

```python
classifier.fit(X_train,y_train)
```

```python
#Predicting the test set result
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
X_test
```

```python
#since it is scaled we cannot directly interpret the result
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
```
